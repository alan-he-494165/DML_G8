```python
import numpy as np
import yfinance as yf

class Ticker_Day:
    def __init__(self, symbol, date = [], high = [], low = [], open = [], close = [], volume = []):
        self.symbol = symbol
        self.high = high
        self.low = low
        self.open = open
        self.close = close
        self.volume = volume
        self.date = date
    

class YFFetcher:
    def __init__(self):
        self.dates_avail = None

    def fetch_data(ticker, start_date, end_date):
        data = yf.download(ticker, start=start_date, end=end_date)
        if data is None or data.empty:
            return None
        ticker_days = Ticker_Day(symbol=ticker)
        ticker_days.date = data.index.tolist()
        ticker_days.high = data['High'].tolist()
        ticker_days.low = data['Low'].tolist()
        ticker_days.open = data['Open'].tolist()
        ticker_days.close = data['Close'].tolist()
        ticker_days.volume = data['Volume'].tolist()
        return ticker_days
    
```
